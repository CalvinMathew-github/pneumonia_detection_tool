```python
import os
import glob
import time
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

from PIL import Image

import torch
import torchvision.utils as vutils
import torchvision.transforms as transforms

from keras import models, layers, optimizers
from keras.preprocessing.image import array_to_img, img_to_array, load_img

import tensorflow as tf

# Dataset paths
base_path = '/kaggle/input/chestxray/xray'
train_path = os.path.join(base_path, 'train/')
test_path = os.path.join(base_path, 'test/')
val_path = os.path.join(base_path, 'val/')

# Fetching image file paths
normal_images = glob.glob(os.path.join(base_path, '**/NORMAL/*'), recursive=True)
pneumonia_images = glob.glob(os.path.join(base_path, '**/PNEUMONIA/*'), recursive=True)

train_normal = glob.glob(os.path.join(train_path, 'NORMAL/*'))
train_pneumonia = glob.glob(os.path.join(train_path, 'PNEUMONIA/*'))

test_normal = os.listdir(os.path.join(test_path, 'NORMAL/'))
test_pneumonia = os.listdir(os.path.join(test_path, 'PNEUMONIA/'))

# Summarizing dataset
print(f'Dataset Summary:')
print(f'Total NORMAL images: {len(normal_images)}')
print(f'Total PNEUMONIA images: {len(pneumonia_images)}')
print(f'Training NORMAL: {len(train_normal)}, PNEUMONIA: {len(train_pneumonia)}')
print(f'Testing NORMAL: {len(test_normal)}, PNEUMONIA: {len(test_pneumonia)}')
print(f'Validation NORMAL: {len(glob.glob(os.path.join(val_path, "NORMAL/*")))}')
print(f'Validation PNEUMONIA: {len(glob.glob(os.path.join(val_path, "PNEUMONIA/*")))}')

# Utility function for calculating elapsed time
def elapsed_time(start_time, end_time):
    total_time = end_time - start_time
    if total_time < 60:
        return f'{total_time:.2f} seconds'
    elif total_time < 3600:
        return f'{total_time//60:.0f} minutes {total_time%60:.2f} seconds'
    else:
        return f'{total_time//3600:.0f} hours {(total_time%3600)//60:.0f} minutes {total_time%60:.2f} seconds'

# Function to load and preprocess images
def load_images(image_paths, image_size=(256, 256), shuffle=True):
    start_time = time.time()
    images = []
    
    for idx, path in enumerate(image_paths):
        if idx % (len(image_paths) // 10) == 0:
            print(f'Processing {idx}/{len(image_paths)} ({100 * idx / len(image_paths):.0f}%)')
        
        img = load_img(path, target_size=image_size, color_mode='grayscale')
        img_array = img_to_array(img)
        images.append(img_array)

    images = np.array(images).astype('float32') / 255.0
    
    if shuffle:
        np.random.shuffle(images)

    print(f'Finished processing in {elapsed_time(start_time, time.time())}')
    return images

# Loading datasets
print('Loading NORMAL images...')
normal_images = load_images(normal_images)
print('Loading PNEUMONIA images...')
pneumonia_images = load_images(pneumonia_images)

# Function to create and display image grid
# Function to create and display image grid
def display_image_grid(images, nrows=4, ncols=5, show=True):
    start_time = time.time()

    num_images = len(images)
    if num_images < nrows * ncols:
        nrows = num_images // ncols
        if num_images % ncols != 0:
            nrows += 1

    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    
    # Ensure images are 3D and in the correct shape (height, width, channels)
    images_tensors = [np.transpose(images[i], (2, 0, 1)) for i in range(nrows * ncols) if i < num_images]
    images_tensors = torch.Tensor(images_tensors).to(device)
    
    grid = np.transpose(
        vutils.make_grid(images_tensors, nrow=ncols, padding=2, normalize=True).cpu(),
        (1, 2, 0)
    )
    
    if show:
        plt.figure(figsize=(12, 12))
        plt.axis("off")
        plt.imshow(grid, cmap='gray')
        plt.show()

    print(f'Image grid created in {elapsed_time(start_time, time.time())}')
    return grid


# Display grids for NORMAL and PNEUMONIA images
normal_grid = display_image_grid(normal_images, show=False)
pneumonia_grid = display_image_grid(pneumonia_images, show=False)

fig, axes = plt.subplots(1, 2, figsize=(19, 8))
axes[0].imshow(normal_grid, cmap='gray')
axes[0].set_title('NORMAL X-Rays', fontsize=27)
axes[0].axis('off')
axes[1].imshow(pneumonia_grid, cmap='gray')
axes[1].set_title('PNEUMONIA X-Rays', fontsize=27)
axes[1].axis('off')
plt.tight_layout()
plt.show()

# GAN hyperparameters
epochs = 50
batch_size = 32
latent_dim = 100
image_size = (256, 256, 1)
learning_rate = 0.0001
beta1 = 0.5

# Discriminator Model
def build_discriminator(input_shape):
    model = models.Sequential([
        layers.Conv2D(128, (5, 5), padding='same', input_shape=input_shape),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Flatten(),
        layers.Dropout(0.4),
        layers.Dense(1, activation='sigmoid')
    ])
    
    opt = optimizers.Adam(lr=learning_rate, beta_1=beta1)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model

# Generator Model
def build_generator(latent_dim):
    model = models.Sequential([
        layers.Dense(128 * 8 * 8, input_dim=latent_dim),
        layers.LeakyReLU(alpha=0.2),
        layers.Reshape((8, 8, 128)),
        layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(1, (5, 5), activation='tanh', padding='same')
    ])
    return model

# GAN Model
def build_gan(generator, discriminator):
    discriminator.trainable = False
    gan_model = models.Sequential([generator, discriminator])
    opt = optimizers.Adam(lr=learning_rate, beta_1=beta1)
    gan_model.compile(loss='binary_crossentropy', optimizer=opt)
    return gan_model

# Function to generate random points in latent space
def generate_latent_points(latent_dim, n_samples):
    return np.random.randn(latent_dim * n_samples).reshape(n_samples, latent_dim)

# Function to generate fake samples
def generate_fake_samples(generator, latent_dim, n_samples):
    z_input = generate_latent_points(latent_dim, n_samples)
    generated = generator.predict(z_input)
    return generated, np.zeros((n_samples, 1))

# Function to select a random sample of real images
def get_real_samples(dataset, n_samples):
    ix = np.random.randint(0, dataset.shape[0], n_samples)
    return dataset[ix], np.ones((n_samples, 1))

# Function to display generated images
def show_generated_images(images, epoch, nrows=4, ncols=5):
    plt.figure(figsize=(10, 10))
    for i in range(nrows * ncols):
        plt.subplot(nrows, ncols, i + 1)
        plt.imshow(images[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.savefig(f'epoch_{epoch+1}.png')
    plt.show()

# Function to evaluate performance
def evaluate_performance(epoch, generator, discriminator, dataset, latent_dim, n_samples=100):
    real_images, real_labels = get_real_samples(dataset, n_samples)
    real_acc = discriminator.evaluate(real_images, real_labels, verbose=0)[1]
    
    fake_images, fake_labels = generate_fake_samples(generator, latent_dim, n_samples)
    fake_acc = discriminator.evaluate(fake_images, fake_labels, verbose=0)[1]
    
    print(f'Epoch {epoch+1} - Real Accuracy: {real_acc:.2f}, Fake Accuracy: {fake_acc:.2f}')
    
    show_generated_images(fake_images, epoch)

# Plotting loss curves
def plot_loss_curves(losses):
    plt.figure(figsize=(10, 5))
    plt.plot(losses[0], label="Discriminator Real Loss")
    plt.plot(losses[1], label="Discriminator Fake Loss")
    plt.plot(losses[2], label="Generator Loss")
    plt.title("Training Loss")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

# Training function
def train_gan(generator, discriminator, gan_model, dataset, latent_dim=100, epochs=100, batch_size=32):
    start_time = time.time()
    
    batch_per_epoch = dataset.shape[0] // batch_size
    half_batch = batch_size // 2
    
    losses = [[], [], []]
    
    print('Training GAN...')
    for epoch in range(epochs):
        for batch in range(batch_per_epoch):
            real_images, real_labels = get_real_samples(dataset, half_batch)
            d_loss_real = discriminator.train_on_batch(real_images, real_labels)
            
            fake_images, fake_labels = generate_fake_samples(generator, latent_dim, half_batch)
            d_loss_fake = discriminator.train_on_batch(fake_images, fake_labels)
            
            z_input = generate_latent_points(latent_dim, batch_size)
            fake_labels = np.ones((batch_size, 1))
            g_loss = gan_model.train_on_batch(z_input, fake_labels)
            
            losses[0].append(d_loss_real[0])
            losses[1].append(d_loss_fake[0])
            losses[2].append(g_loss)
        
        print(f'Epoch {epoch+1}/{epochs} - Discriminator Real Loss: {d_loss_real[0]:.3f}, Fake Loss: {d_loss_fake[0]:.3f}, Generator Loss: {g_loss:.3f}')
        
        if (epoch + 1) % (epochs // 10) == 0:
            evaluate_performance(epoch, generator, discriminator, dataset, latent_dim)
    
    print(f'Training completed in {elapsed_time(start_time, time.time())}')
    plot_loss_curves(losses)

# Build and train the GAN
discriminator = build_discriminator(image_size)
generator = build_generator(latent_dim)
gan = build_gan(generator, discriminator)

train_gan(generator, discriminator, gan, normal_images, latent_dim, epochs, batch_size)
```

```python
import os
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import numpy as np
import torch
import torch.nn.functional as F
from skimage.metrics import structural_similarity as ssim
from torchvision.models import inception_v3
from scipy.stats import entropy
from scipy.linalg import sqrtm

device = torch.device("cpu")


class CustomImageDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.transform = transform
        self.image_paths = [os.path.join(image_dir, fname) for fname in os.listdir(image_dir) if fname.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff', '.webp'))]

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image

def load_images(image_dir, batch_size=16, image_size=(256, 256)):  
    transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5], std=[0.5]) 
    ])
    
    dataset = CustomImageDataset(image_dir, transform=transform)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)
    
    return dataloader

def images_to_numpy(dataloader):
    images = []
    for batch in dataloader:
        images.append(batch)
    images = torch.cat(images)
    images = images.permute(0, 2, 3, 1)  
    images = images.numpy()
    images = (images * 255).astype(np.uint8)  
    return images


def inception_score(images, batch_size=16, splits=10):  
    N = len(images)
    dtype = torch.cuda.FloatTensor if torch.cuda.is_available() else torch.FloatTensor
    inception_model = inception_v3(pretrained=True, transform_input=False).to(device).eval()
    
    def get_pred(x):
        with torch.no_grad():
            x = x.to(device)
            x = F.interpolate(x, size=(299, 299), mode='bilinear', align_corners=False)
            x = inception_model(x)
        return F.softmax(x, dim=1).data.cpu().numpy()

    preds = np.zeros((N, 1000))
    
    for i in range(0, N, batch_size):
        batch = images[i:i + batch_size]
        if batch.size(0) == 0:
            continue
        preds[i:i + batch_size] = get_pred(batch)
        
    split_scores = []
    for k in range(splits):
        part = preds[k * (N // splits): (k + 1) * (N // splits), :]
        py = np.mean(part, axis=0)
        scores = []
        for i in range(part.shape[0]):
            pyx = part[i, :]
            scores.append(entropy(pyx, py))
        split_scores.append(np.exp(np.mean(scores)))
        
    return np.mean(split_scores), np.std(split_scores)

def calculate_ssim(real_images, generated_images):
    ssim_values = []
    for real, gen in zip(real_images, generated_images):
        
        min_dim = min(real.shape[0], real.shape[1])
        win_size = min(min_dim, 7)
        
        
        if win_size % 2 == 0:
            win_size -= 1
        
        
        ssim_value = ssim(real, gen, data_range=real.max() - real.min(), win_size=win_size, multichannel=True)
        ssim_values.append(ssim_value)
        
    return np.mean(ssim_values)

# Load real and generated images
real_dataloader = load_images('/kaggle/input/chestxray/xray/train/NORMAL')
generated_dataloader = load_images('/kaggle/input/generated')

real_images = images_to_numpy(real_dataloader)
generated_images = images_to_numpy(generated_dataloader)


real_images = torch.tensor(real_images).permute(0, 3, 1, 2).float() / 255.0
generated_images = torch.tensor(generated_images).permute(0, 3, 1, 2).float() / 255.0

real_images = real_images.to(device)
generated_images = generated_images.to(device)

inception_mean, inception_std = inception_score(generated_images)
print(f"Inception Score: Mean = {inception_mean}, Std = {inception_std}")

mean_ssim = calculate_ssim(real_images.cpu().permute(0, 2, 3, 1).numpy(), generated_images.cpu().permute(0, 2, 3, 1).numpy())
print(f"Mean SSIM: {mean_ssim}")
```
